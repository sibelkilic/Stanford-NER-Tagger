```python
import os
from nltk.tag import StanfordNERTagger

JAVA_PATH = r'C:\Program Files (x86)\Java\jre1.8.0_241\bin\java.exe'
os.environ['JAVAHOME'] = JAVA_PATH

STANFORD_CLASSIFIER_PATH = 'C:/.../Desktop/stanford-ner-2014-08-27/classifiers/english.muc.7class.distsim.crf.ser.gz'
STANFORD_NER_JAR_PATH = 'C:/.../Desktop/stanford-ner-2014-08-27/stanford-ner.jar'

sn = StanfordNERTagger(STANFORD_CLASSIFIER_PATH,
                       path_to_jar=STANFORD_NER_JAR_PATH)
sn
```

```python
currentDT = datetime.datetime.now()
print (str(currentDT))

text_enc = textnew.encode('ascii', errors='ignore').decode('utf-8')
ner_tagged = sn.tag(text_enc.split())

currentDT = datetime.datetime.now()
print (str(currentDT))

print(ner_tagged)
```

```python
currentDT = datetime.datetime.now()
print (str(currentDT))

named_entities = []
temp_entity_name = ''
temp_named_entity = None
for term, tag in ner_tagged:
    if tag != 'O':
        temp_entity_name = ' '.join([temp_entity_name, term]).strip()
        temp_named_entity = (temp_entity_name, tag)
    else:
        if temp_named_entity:
            named_entities.append(temp_named_entity)
            temp_entity_name = ''
            temp_named_entity = None
            
currentDT = datetime.datetime.now()
print (str(currentDT))            
```
